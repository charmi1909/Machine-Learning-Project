```python
import pandas as pd

df = pd.read_csv('../data/pre_processed_data.csv')

print(df.head())

print("Dataset shape:", df.shape)

df.isnull().sum()
```

```python
X_df = df.drop('fraud_reported', axis=1)  

y = df['fraud_reported'].values.reshape(-1, 1)  

print("Features shape:", X_df.shape)
print("Target shape:", y.shape)
print("First 5 rows of features:\n", X_df.head())
print("First 5 values of target:\n", y[:5])
```

```python
X_df = X_df.apply(pd.to_numeric, errors='coerce')

X_df = X_df.fillna(0)

X = X_df.values

print("First 5 rows of numeric features:\n", X[:5])
```

```python
import numpy as np

X = (X - X.mean(axis=0)) / (X.std(axis=0) + 1e-8)  

print("First 5 rows of scaled features:\n", X[:5])
```

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

test_values = np.array([-2, 0, 2])
print("Sigmoid outputs:", sigmoid(test_values))
```

```python
n_features = X.shape[1]

weights = np.zeros((n_features, 1))

bias = 0

learning_rate = 0.01
n_iterations = 1000

print("Weights shape:", weights.shape)
print("Bias:", bias)
```

```python
for i in range(n_iterations):
    z = np.dot(X, weights) + bias
    
    y_pred = sigmoid(z)
    
    dz = y_pred - y                     
    dw = np.dot(X.T, dz) / len(y)       
    db = np.sum(dz) / len(y)           
    
    weights -= learning_rate * dw
    bias -= learning_rate * db
    
    if i % 100 == 0:
        loss = -np.mean(y * np.log(y_pred + 1e-15) + (1 - y) * np.log(1 - y_pred + 1e-15))
        print(f"Iteration {i}, Loss: {loss}")
```

```python
def predict(X, weights, bias, threshold=0.5):
    y_pred_prob = sigmoid(np.dot(X, weights) + bias)
    return (y_pred_prob >= threshold).astype(int)

y_pred_class = predict(X, weights, bias)

print("First 10 predicted classes:", y_pred_class[:10].flatten())
```

```python
accuracy = np.mean(y_pred_class == y)

print(f"Training Accuracy: {accuracy * 100:.2f}%")
```
